<p>In this notebook, we will take a look at a dataset of The Office episodes, and try to understand how the popularity and quality of the series varied over time.
```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the CSV as a DataFrame
office_df = pd.read_csv('datasets/office_episodes.csv')
office_df.head()
```

```python
# create a color scheme
colors = []

for lab, rows in office_df.iterrows():
    if rows['scaled_ratings'] < 0.25:
        colors.append('red')
    elif rows['scaled_ratings'] < 0.50:
        colors.append('orange')
    elif rows['scaled_ratings'] < 0.75:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')

colors[:10]
```

```python
# create a sizing system
sizes = []

for lab, rows in office_df.iterrows():
    if rows['has_guests'] == True:
        sizes.append(250)
    else:
        sizes.append(25)
        
sizes[:10]
```

```python
# create a scatter plot
fig = plt.figure(figsize=(12, 8))

plt.scatter(office_df['episode_number'], office_df['viewership_mil'], c=colors, s=sizes)
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')

plt.show()
```
